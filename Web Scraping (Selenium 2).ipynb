```python
#Question 2
Brand_Names = []
Product_Names=[]
Price=[]
Return=[]
Expected_Delivery=[]
Availabilty=[]
Product_Url=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    
#brand names 
try:
    brand=driver.find_elements_by_xpath('//*[@id="productOverview_feature_div"]/div/table/tbody/tr[3]/td[2]/span')
    Brand_Names.append(brand.text.replace("/n","new line"))
except:
    Brand_Names.append('_')
    
    
#product names 
try:
    product=driver.find_elements_by_xpath('//*[@id="title"]')
    Product_Names.append(product.text)
except:
    Product_Names.append('_')
    
#price
try:
    price=driver.find_elements_by_xpath('//*[@id="corePrice_desktop"]/div/table/tbody/tr[2]/td[2]/span[1]')
    Price.append(price.text)
except:
    Price.append('_')
    
#return    
try:
    return_=driver.find_elements_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a')
    Return.append(return_.text)
except:
    Return.append('_')
    
#expected delivery
try:
    expected=driver.find_elements_by_xpath('//*[@id="AMAZON_DELIVERED"]/div[2]/a')
    Expected_Delivery.append(expected.text)
except:
    Expected_Delivery.append('_')
    
#availability
try:
    availabilty=driver.find_elements_by_xpath('//*[@id="ddmDeliveryMessage"]/b')
    Availabilty.append(availabilty.text)
except:
    Availabilty.append('_')
    
#Product Url
try:
    product_url=driver.find_elements_by_xpath('//*[@id="productTitle"]')
    Product_Url.append(product_url.text)
except:
    Product_Url.append('_')
```

```python
print(len(Brand_Names))
print(len(Product_Names))
print(len(Price))
print(len(Return))
print(len(Expected_Delivery))
print(len(Availabilty))
print(len(Product_Url))
```
